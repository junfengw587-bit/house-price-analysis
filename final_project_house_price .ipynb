```python
# 导入绘图库
import matplotlib.pyplot as plt
import seaborn as sns
# 导入机器学习相关库
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.pipeline import Pipeline

# 解决绘图中文乱码问题
plt.rcParams['font.sans-serif'] = ['SimHei', 'Arial Unicode MS']
plt.rcParams['axes.unicode_minus'] = False

# 忽略无关警告
import warnings
warnings.filterwarnings('ignore')
```

```python
# 1.1 输出每一列的数据类型
print("===== 1.1 每列的数据类型 =====")
df.dtypes
```

```python
# 1.2 删除id和Unnamed: 0列，inplace=True直接修改原数据
df.drop(columns=['id', 'Unnamed: 0'], inplace=True, errors='ignore')

# 输出数据的统计描述信息
print("===== 1.2 数据统计描述 =====")
df.describe()
```

```python
# 1.3 统计楼层数的出现次数，转为规范的表格格式
print("===== 1.3 楼层数分布统计 =====")
df['floors'].value_counts().to_frame()
```

```python
# 1.4 箱线图对比有无水滨景观的房价分布与异常值
plt.figure(figsize=(10, 6))
sns.boxplot(x='waterfront', y='price', data=df)
plt.title('有无水滨景观的房屋价格对比')
plt.xlabel('是否有水滨景观（0=无，1=有）')
plt.ylabel('房屋价格')
plt.show()
```

```python
# 1.5 回归图查看房屋地上面积与房价的相关性
plt.figure(figsize=(10, 6))
sns.regplot(x='sqft_above', y='price', data=df, line_kws={'color':'red'})
plt.title('房屋地上面积与房价的相关性')
plt.xlabel('房屋地上面积（平方英尺）')
plt.ylabel('房屋价格')
plt.show()
```

```python
# 1.6 用居住面积sqft_living做单特征线性回归
# 定义特征X（必须是二维格式，所以用双括号）和目标变量y（房价）
X_single = df[['sqft_living']]
y = df['price']

# 创建并拟合线性回归模型
lr_single = LinearRegression()
lr_single.fit(X_single, y)

# 计算并输出R²值
r2_single = lr_single.score(X_single, y)
print(f"===== 1.6 单特征线性回归结果 =====")
print(f"R²值：{r2_single:.4f}")
```

```python
# 1.7 用多个房屋核心特征做线性回归
multi_features = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 
                  'floors', 'waterfront', 'view', 'condition', 'grade', 
                  'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 
                  'zipcode', 'lat', 'long']

# 定义多特征X和目标变量y（和1.6的y完全一致，无需重复定义）
X_multi = df[multi_features]

# 创建并拟合多特征线性回归模型
lr_multi = LinearRegression()
lr_multi.fit(X_multi, y)

# 输出R²值
r2_multi = lr_multi.score(X_multi, y)
print(f"===== 1.7 多特征线性回归结果 =====")
print(f"R²值：{r2_multi:.4f}")
```

```python
# 1.8 用Pipeline管道构建多特征预测模型
# 管道将数据标准化和线性回归打包成一个整体，保证数据处理流程的一致性，是机器学习的工程化规范写法
pipe_model = Pipeline([
    ('scaler', StandardScaler()),  # 第一步：对特征做标准化处理
    ('linear_reg', LinearRegression())  # 第二步：线性回归建模
])

# 拟合管道模型
pipe_model.fit(X_multi, y)

# 计算并输出模型的R²值
r2_pipe = pipe_model.score(X_multi, y)
print(f"===== 1.8 多特征管道模型结果 =====")
print(f"R²值：{r2_pipe:.4f}")
```

```python
# 1.9 Ridge回归模型（设置alpha=0.1）
# 第一步：将数据集拆分为训练集（80%，用于训练模型）和测试集（20%，用于评估模型在新数据上的表现）
X_train, X_test, y_train, y_test = train_test_split(
    X_multi, y, test_size=0.2, random_state=42
)

# 创建Ridge回归模型，设置alpha=0.1
ridge_model = Ridge(alpha=0.1)

# 用训练集拟合模型
ridge_model.fit(X_train, y_train)

# 用测试集计算R²，评估模型的泛化能力
r2_ridge = ridge_model.score(X_test, y_test)
print(f"===== 1.9 Ridge回归模型结果 =====")
print(f"测试集R²值：{r2_ridge:.4f}")
```

```python
# 1.10 二次多项式变换+Ridge回归（设置alpha=0.1）
# 用管道将多项式变换、标准化、Ridge回归打包，保证流程规范
poly_pipe = Pipeline([
    ('poly', PolynomialFeatures(degree=2)),  # 第一步：二次多项式变换，捕捉特征间的非线性关系
    ('scaler', StandardScaler()),  # 第二步：数据标准化
    ('ridge', Ridge(alpha=0.1))  # 第三步：Ridge回归，alpha=0.1
])

# 用训练集拟合模型
poly_pipe.fit(X_train, y_train)

# 用测试集计算R²，评估模型效果
r2_poly = poly_pipe.score(X_test, y_test)
print(f"===== 1.10 二次多项式+Ridge回归结果 =====")
print(f"测试集R²值：{r2_poly:.4f}")
```
